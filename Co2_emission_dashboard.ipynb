```python
target_correlation = emission_source.corr()['total_emission'].abs()
```

```python
target_correlation.sort_values(ascending=False)
```

```python
# Filter columns where the correlation with 'total_emission' is 0.5 or higher
strong_correlation_columns = target_correlation[target_correlation >= 0.5].index

# Create a new DataFrame with only columns that have a correlation >= 0.5
emission_source_filtered = emission_source[strong_correlation_columns]

```

```python
emission_source_filtered_corr = pd.DataFrame(emission_source_filtered.corr())
```

```python
print(emission_source_filtered_corr)
```

```python
sns.heatmap(emission_source_filtered.corr())
```

The emission source below are the sources considered for the impact on total emission, the other emission sources have been dropped.

```python
emission_source_filtered.corr()['total_emission'].sort_values(key = abs, ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select top variables for visualization
top_vars = emission_source_filtered_corr.index[1:11]  # Exclude total_emission itself, select top 10 variables
top_correlation_matrix = df[top_vars].join(df['total_emission']).corr()

# Plot heatmap for top correlations with total_emission
plt.figure(figsize=(10, 8))
sns.heatmap(top_correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Top Correlated Variables with Total Emission")
plt.show()
```

```python
def normalize(df):
    normalize_res = (df - df.max()) / (df.max() - df.min())
    return normalize_res

temp_emission = df.groupby("Year").agg({"Average Temperature Â°C": "mean", "total_emission": "mean", "Urban population": "mean"})
norm_temp = normalize(temp_emission)


norm_temp.plot(figsize=(20, 6))
plt.title("CO2 Emission & Temperature")
plt.show()
```

```python
import pandas as pd
import pycountry
import pycountry_convert as pc

# List of countries
countries = [
    'Afghanistan', 'Albania', 'Algeria', 'American Samoa', 'Andorra',
    'Angola', 'Anguilla', 'Antigua and Barbuda', 'Argentina',
    'Armenia', 'Aruba', 'Australia', 'Austria', 'Azerbaijan',
    'Bahamas', 'Bahrain', 'Bangladesh', 'Barbados', 'Belarus',
    'Belgium', 'Belgium-Luxembourg', 'Belize', 'Benin', 'Bermuda',
    'Bhutan', 'Bolivia (Plurinational State of)', 'Bosnia and Herzegovina',
    'Botswana', 'Brazil', 'British Virgin Islands', 'Brunei Darussalam',
    'Bulgaria', 'Burkina Faso', 'Burundi', 'Cabo Verde', 'Cambodia',
    'Cameroon', 'Canada', 'Cayman Islands', 'Central African Republic',
    'Chad', 'Channel Islands', 'Chile', 'China', 'China, Hong Kong SAR',
    # ... (rest of the list)
]

# Function to map country to continent
def get_continent(country_name):
    try:
        country_code = pycountry.countries.lookup(country_name).alpha_2
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except (KeyError, LookupError):
        return 'Unknown'

# Create a DataFrame
df_country = pd.DataFrame(countries, columns=['Country'])

# Apply continent mapping
df_country['Continent'] = df_country['Country'].apply(get_continent)
df_country['total_population'] = df['Total Population - Male'] + df['Total Population - Female']

# Display the results
print(df_country)
```

```python
continent_colors = {
    'Asia': 'red',
    'Europe': 'blue',
    'Africa': 'green',
    'North America': 'orange',
    'South America': 'purple',
    'Oceania': 'cyan',
    'Antarctica': 'gray',
    'Unknown': 'black'  # For any unknown continents
}
df['Color'] = df_country['Continent'].map(continent_colors)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrame (replace this with your actual DataFrame)
# df = pd.DataFrame(...)

# Print unique values in the 'continent' column to diagnose the issue
print("Unique values in 'continent':", df['continent'].unique())

# Define a color mapping for each continent
continent_colors = {
    'Asia': 'red',
    'Africa': 'blue',
    'Europe': 'green',
    'North America': 'purple',
    'South America': 'orange',
    'Oceania': 'brown',
    'Antarctica': 'cyan'
}

# Map the continents to colors
df['continent_color'] = df['continent'].map(continent_colors)

# Check if any continent color is NaN
print("Count of NaN values in 'continent_color':", df['continent_color'].isna().sum())

# Replace NaN values with a default color, e.g., 'grey'
df['continent_color'].fillna('grey', inplace=True)

# Plot the scatter plot using the color mapping
plt.scatter(df['Average Temperature Â°C'], df['total_emission'],
            s=df["total_population"],
            c=df['continent_color'],
            alpha=0.7)

# Add labels and title
plt.xlabel('Average Temperature Â°C')
plt.ylabel('Total Emission')
plt.title('Total Emission vs Average Temperature by Continent')
plt.figsize=(20,10)

# Show the plot
plt.show()
```
